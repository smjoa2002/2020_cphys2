```python
import os
import glob
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
IMAGE_DIR="../input/chinese-mnist/data/data"
TRAIN_DIR="../input/chinese-mnist/data/data"
TEST_DIR="../input/chinese-mnist/data/data"

def load_data():
    file_list=glob.glob(IMAGE_DIR+"/*jpg")
    x=[]
    y=[]
    for fname in file_list:
        with Image.open(fname) as img:
            np_img=np.array(img)
        label=int(os.path.split(fname)[-1].split('.')[0].split('_')[3])-1
        
        x.append(np_img)
        y.append(label)
    x,y=np.array(x),np.array(y)
    x=np.expand_dims(x,-1)
    x=x/255
    return x,y
X,y=load_data()
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33333,random_state=1)
```

```python
idx=0
for i in X_train:
    xim=i.squeeze(2)
    xim=xim*255
    img=Image.fromarray(xim)
    img=img.convert("L")
    fn=idx
    img.save('..\input\cbnu-comphys2020\train_fig\train_fig\fig{:04d}.jpg'.format(fn))
    idx +=1
    
```
